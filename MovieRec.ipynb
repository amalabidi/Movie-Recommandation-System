```python
import pandas as pd 
import matplotlib as plt
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=pd.read_csv('movies.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
selected_features=['genres','keywords','tagline','cast','director','title']
```

```python
for feature in selected_features:
   data[feature]=data[feature].fillna('')
```

```python
#combining the features
features=data['genres']+''+data['keywords']+''+data['tagline']+''+data['cast']+''+data['director']+''+data['title']
```

```python
print(features)
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(features)
print(feature_vectors)
```

```python
similarity=cosine_similarity(feature_vectors)
print(similarity)
```

```python
movie_name=input('enter the movie name')
```

```python
list_titles=data['title'].tolist()
```

```python
close_match_list=difflib.get_close_matches(movie_name,list_titles)
print(close_match_list)
```

```python
#finding the index of the closest match 
close_match=close_match_list[0]
index=data[data.title == close_match]['index'].values[0]
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index]))
print(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = data[data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
